```python
import numpy as np
import pandas as pd
from scipy.misc import imread
import matplotlib.pyplot as plt
import seaborn as sns

path = "../NetDissect-Lite/dataset/broden1_224/"
```

```python
index = pd.read_csv(path + "index.csv")
index.head()
```

```python
def findUnique(path):
    base_path = "../NetDissect-Lite/dataset/broden1_224/images/"
    if (type(path) is not str):
        return np.nan
    else: 
        image = imread(base_path + path)
        return np.unique(image)[1:] # do not include 0
        
```

```python
def countUniqueConcepts(concept): 
    new_index = pd.read_csv("../NetDissect-Lite/dataset/broden1_224/index.csv")
    
    # create columns "n_concepts" and concepts
    new_index[concept + "s"] = new_index[concept].apply(lambda x: findUnique(x))# findUnique(x[concept]))
    # new_index["n_" + concept + "s"] = new_index[concept + "s"].apply(lambda x: np.nan if (type(x) is float) else len(x)-1)
    

    return new_index
```

```python
new_index = countUniqueConcepts("object")
```

```python
new_index["n_" + concept + "s"] = new_index[concept + "s"].apply(lambda x: 0 if (type(x) is float) else max(0, len(x)-1))
new_index.head(10)
```

```python
new_index.to_csv("../NetDissect-Lite/dataset/broden1_224/" + "new_index_" + concept + ".csv")
```

```python
index_objects = new_index[["image", "split", "object", "objects", "n_objects"]]
index_objects.describe()
```

```python
object_counts = new_index[new_index["n_objects"] > 0]["n_objects"].values
object_counts.sort()
sns.distplot(object_counts);
```
